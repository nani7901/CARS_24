## <b>Conactenating the Multiple Data Frames in to single</b>

```python
#combining all those in to a single Dataframe
Cars_24=pd.concat([NewDelhi_cars, Mumbai_cars, Hyderabad_cars, Banglore_cars], ignore_index=True)
```

## <b>Sorting the Concatenated DataFrame by Year</b>

```python
# Ensure YOM column is of numeric type for proper sorting
Cars_24['YOM'] = Cars_24['YOM'].astype(int)

# Sort the DataFrame by YOM in ascending order (change to descending by setting ascending=False)
Cars_24 = Cars_24.sort_values(by='YOM', ascending=True).reset_index(drop=True)

# Display the sorted DataFrame
print(Cars_24)
```

##### Why Use the Mean for Imputation?

The mean is used as it reflects the central tendency of Kms_Driven for each brand, preserving brand-specific trends while minimizing distortion in the data distribution. It’s effective when the data is not heavily skewed by outliers.

```python
# Impute missing 'kms_driven' with the mean of each Brand
Cars_24['Kms_Driven'] = Cars_24.groupby('Brand')['Kms_Driven'].transform(lambda x: x.fillna(x.mean()))

# Check for missing values after imputation
print(Cars_24['Kms_Driven'].isnull().sum())
```

##### Replacing Empty Strings and Imputing Missing 'Owner_Type' with Mode

Empty strings in the Owner_Type column are replaced with NaN to standardize missing values. These are then imputed using the mode (most frequent value) to ensure consistency, as the mode represents the most common ownership type in the dataset.

```python
# Replace empty strings or whitespace with NaN
Cars_24['Owner_Type'].replace('', np.nan, inplace=True)
Cars_24['Owner_Type'] = Cars_24['Owner_Type'].apply(lambda x: np.nan if isinstance(x, str) and x.strip() == '' else x)

# Now impute missing values with the mode (most frequent value)
owner_mode = Cars_24['Owner_Type'].mode()[0]  # Get the most frequent value
Cars_24['Owner_Type'] = Cars_24['Owner_Type'].fillna(owner_mode)  # Fill missing with the mode

# Check if missing values have been filled
print(Cars_24['Owner_Type'].isnull().sum())  # Should print 0 if no missing values remain
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Get the unique years in the dataset
unique_years = df['YOM'].unique()

# Step 2: Set up the plot
plt.figure(figsize=(14, 8))

# Step 3: Generate KDE plots for each year
for i, year in enumerate(unique_years):
    # Filter the data for the specific year
    year_data = df[df['YOM'] == year]
    
    # Create the KDE plot with the 'viridis' colormap for each plot
    color = plt.cm.viridis(i / len(unique_years))  # Map the color using the colormap
    
    sns.kdeplot(year_data['Price_in_Lakhs'], fill=True, color=color, linewidth=2, label=str(year))

# Add title and labels
plt.title('KDE Plot for Price in Lakhs by Year of Manufacture', fontsize=16)
plt.xlabel('Price in Lakhs', fontsize=12)
plt.ylabel('Density', fontsize=12)

# Add a legend
plt.legend(title='Year of Manufacture')

# Display the plot
plt.tight_layout()
plt.show()
```

#### Insights from the KDE Plot:

The KDE Plot reveals the distribution of car prices (in lakhs) by year of manufacture:

Price Trend: Cars from more recent years tend to have higher prices, as seen by the density curves shifting toward higher values over time.

Outliers: Some older cars (2011-2013) show unexpectedly high prices (around 3 lakhs), which appear as outliers in the plot.

Price Variation: While there’s slight variation in price distribution within each year, the overall trend of increasing prices with newer cars remains consistent.

In summary, the analysis highlights that newer cars generally command higher prices, with occasional outliers in the older model range.

```python
# Step 1: Group by Brand and Year of Manufacture (YOM) and count the number of cars
brand_year_counts = df.groupby(['Brand', 'YOM']).size().reset_index(name='Car Count')

# Step 2: Set up subplots (2x2 grid)
fig, axes = plt.subplots(2, 2, figsize=(14, 12))

# Plot 1: Bar plot for the number of cars by Brand and YOM
sns.barplot(x='YOM', y='Car Count', hue='Brand', data=brand_year_counts, ax=axes[0, 0])
axes[0, 0].set_title('Car Count by Brand and Year of Manufacture')
axes[0, 0].set_xlabel('Year of Manufacture')
axes[0, 0].set_ylabel('Car Count')

# Plot 2: Line plot to show trends for each Brand over the years
sns.lineplot(x='YOM', y='Car Count', hue='Brand', data=brand_year_counts, marker='o', ax=axes[0, 1])
axes[0, 1].set_title('Trend of Cars by Brand Over the Years')
axes[0, 1].set_xlabel('Year of Manufacture')
axes[0, 1].set_ylabel('Car Count')

# Plot 3: Stacked bar plot for the number of cars by Brand and Year of Manufacture
# Pivot the data to get the count of cars for each brand by year
pivot_table = brand_year_counts.pivot_table(index='YOM', columns='Brand', values='Car Count', fill_value=0)

# Create a stacked bar plot
pivot_table.plot(kind='bar', stacked=True, ax=axes[1, 0], colormap='viridis')
axes[1, 0].set_title('Stacked Bar Plot of Cars by Brand and Year')
axes[1, 0].set_xlabel('Year of Manufacture')
axes[1, 0].set_ylabel('Car Count')

# Plot 4: Boxplot to show the distribution of car counts by Brand
sns.boxplot(x='Brand', y='Car Count', data=brand_year_counts, ax=axes[1, 1])
axes[1, 1].set_title('Distribution of Car Count by Brand')
axes[1, 1].set_xlabel('Brand')
axes[1, 1].set_ylabel('Car Count')

# Rotate x-axis labels to prevent overlap
for ax in axes.flat:
    for label in ax.get_xticklabels():
        label.set_rotation(90)

# Tight layout to ensure plots don't overlap
plt.tight_layout()

# Display the plots
plt.show()
```
